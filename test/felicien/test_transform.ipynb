```python
import pandas as pd
import json

json_file_path = '/Users/feliciendehertogh/Desktop/orange-competitor-analysis/data/products.json'

with open(json_file_path, 'r') as f:
    data_dict = json.load(f)

df = pd.DataFrame(data_dict['products'])

df.head()
```

```python
df[df['download_speed'].notna()]
```

```python
import re
def convert_speed(speed_str):
    if pd.isnull(speed_str):
        return None
    match = re.match(r'(\d+)(mbps|gbps)', speed_str)
    if not match:
        return None
    value, unit = match.groups()
    value = int(value)
    if unit == "gbps":
        value *= 1000
    return value

df['upload_speed'] = df['upload_speed'].apply(convert_speed)
df['upload_speed']
```
